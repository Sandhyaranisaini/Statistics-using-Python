```python
#Task2:
import pandas as pd
import scipy.stats as stats
# Separate the two groups based on whether the computer has a CD player or not
prices_with_cd = computers_data[computers_data['cd'] == 'yes']['price']
prices_without_cd = computers_data[computers_data['cd'] == 'no']['price']

# Perform an independent two-sample t-test (Welch's t-test assuming unequal variances)
t_stat, p_value = stats.ttest_ind(prices_with_cd, prices_without_cd, equal_var=False)

# Output the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Decision rule based on p-value
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in the average prices.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the average prices.")
```

```python
#Task3.a:
import pandas as pd
import scipy.stats as stats
import numpy as np
# Filter premium computers
premium_computers = computers_data[computers_data['premium'] == 'yes']

# Randomly select 25 premium computers
sample_premium = premium_computers['price'].sample(n=25, random_state=1)

# Define the population mean
population_mean = 2200

# Calculate the sample mean and sample standard deviation
sample_mean = np.mean(sample_premium)
sample_std = np.std(sample_premium, ddof=1)  # ddof=1 for sample standard deviation

# Perform the one-sample t-test
t_stat, p_value = stats.ttest_1samp(sample_premium, population_mean)

# Output the results
print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Decision rule based on p-value
if p_value < 0.05:
    print("Reject the null hypothesis: The mean price of premium computers differs significantly from $2200.")
else:
    print("Fail to reject the null hypothesis: The mean price of premium computers does not differ significantly from $2200.")
```

```python
#task3.b:
import pandas as pd
import scipy.stats as stats

# Assuming df is your DataFrame containing the dataset
# Filter the data to separate prices of premium and non-premium computers
premium_computers = computers_data[computers_data['premium'] == 'yes']['price']
non_premium_computers = computers_data[computers_data['premium'] == 'no']['price']

# Perform a two-sample t-test assuming equal variances
t_stat, p_value = stats.ttest_ind(premium_computers, non_premium_computers, equal_var=True)

# Output the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Decision rule based on p-value
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in the mean prices of premium and non-premium computers.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the mean prices of premium and non-premium computers.")
```
