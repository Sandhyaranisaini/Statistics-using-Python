```python
import pandas as pd
computers_data=pd.read_csv('/content/DS1_C8_Computers_Data_Project.csv')
computers_data
```

```python
print(computers_data.isnull().sum())
print(computers_data.dtypes)
```

```python
#TASK 1.a: Analyze purchase likelihood based on specifications
# Step 1: Filter for computers that meet the desired features
filtered_data = computers_data[(computers_data['ram'] >= 8) &  # RAM >= 8GB
                               (computers_data['screen'] >= 14) &  # Screen size >= 14 inches
                               (computers_data['speed'] > 33)]  # Speed > 33 MHz

# Step 2: Calculate the total number of computers in the dataset
total_computers = len(computers_data)

# Step 3: Calculate the number of computers that meet the desired criteria
matching_computers = len(filtered_data)

# Step 4: Calculate the probability (likelihood of a computer having the desired features)
probability = matching_computers / total_computers

# Output the result
print(f"The probability of a computer having at least 8GB of RAM, "
      f"a screen size of at least 14 inches, and a processor speed exceeding 33 MHz is: {probability:.4f}")
```

```python
#TASK1.b:Probability of selling premium computers
#1b.1
# Step 1: Count the number of premium computers
premium_computers = computers_data[computers_data['premium'] == 'yes']

# Step 2: Count the total number of computers in the dataset
total_computers = len(computers_data)

# Step 3: Calculate the probability of selling a premium computer
probability_premium = len(premium_computers) / total_computers

# Output the result
print(f"The probability of selling a premium computer is: {probability_premium:.4f}")
```

```python
#1b.2
# Step 1: Filter for premium computers
premium_computers = computers_data[computers_data['premium'] == 'yes']

# Step 2: Filter for premium computers that have a CD player
premium_with_cd = premium_computers[premium_computers['cd'] == 'yes']

# Step 3: Calculate the probability of a premium computer having a CD player
probability_cd_given_premium = len(premium_with_cd) / len(premium_computers)

# Output the result
print(f"Given that a computer is premium, the probability that it has a CD player is: {probability_cd_given_premium:.4f}")
```

```python
#TASK1.c:
#1c.1:
# Step 1: Filter for computers that are premium and have a screen size greater than 15 inches
premium_with_large_screen = computers_data[(computers_data['premium'] == 'yes') &
                                           (computers_data['screen'] > 15)]

# Step 2: Count the total number of computers in the dataset
total_computers = len(computers_data)

# Step 3: Calculate the probability
probability_premium_large_screen = len(premium_with_large_screen) / total_computers

# Output the result
print(f"The probability of a computer having a screen size greater than 15 inches and being premium is: {probability_premium_large_screen:.4f}")
```

```python
#1c.2:
# Step 1: Filter for non-premium computers
non_premium_computers = computers_data[computers_data['premium'] == 'no']

# Step 2: Filter for non-premium computers with a screen size less than or equal to 14 inches
non_premium_small_screen = non_premium_computers[non_premium_computers['screen'] <= 14]

# Step 3: Calculate the probability
probability_small_screen_given_non_premium = len(non_premium_small_screen) / len(non_premium_computers)

# Output the result
print(f"Given that a computer is not premium, the probability that it has a screen size less than or equal to 14 inches is: {probability_small_screen_given_non_premium:.4f}")
```

```python
#1.d:
#1d.1:
# Step 1: Filter for computers with both a CD player and a multimedia kit
computers_with_cd_and_multimedia = computers_data[(computers_data['cd'] == 'yes') & (computers_data['multi'] == 'yes')]

# Step 2: Count the total number of computers in the dataset
total_computers = len(computers_data)

# Step 3: Calculate the joint probability
joint_probability_cd_and_multimedia = len(computers_with_cd_and_multimedia) / total_computers

# Output the result
print(f"The joint probability of a computer having both a CD player and a multimedia kit is: {joint_probability_cd_and_multimedia:.4f}")
```

```python
#1d.2:
# Step 1: Filter for computers with a CD player
computers_with_cd = computers_data[computers_data['cd'] == 'yes']

# Step 2: Filter for computers with both a CD player and a multimedia kit
computers_with_cd_and_multimedia = computers_with_cd[computers_with_cd['multi'] == 'yes']

# Step 3: Calculate the conditional probability
conditional_probability_multimedia_given_cd = len(computers_with_cd_and_multimedia) / len(computers_with_cd)

# Output the result
print(f"Given that a computer has a CD player, the probability that it also has a multimedia kit is: {conditional_probability_multimedia_given_cd:.4f}")
```

```python
#TASK 2.a:
# Calculate the average price of computers in the dataset
average_price = computers_data['price'].mean()

# Output the result
print(f"The average price of computers in the dataset is: {average_price:.2f}")
```

```python
#TASK 2.b:
import pandas as pd
import numpy as np

# Set the number of samples to draw
sample_size = 100

# 1. Simple Random Sampling (SRS)
random_sample = computers_data['price'].sample(n=sample_size, random_state=1)
average_price_random = random_sample.mean()

# 2. Stratified Sampling (Assuming we stratify based on the 'premium' column)
strata = computers_data.groupby('premium')
sample_strata = strata.apply(lambda x: x['price'].sample(n=sample_size // len(strata), random_state=1)).reset_index(drop=True)
average_price_stratified = sample_strata.mean()

# 3. Systematic Sampling (Take every M-th item)
M = len(computers_data) // sample_size  # Choose every k-th item
systematic_sample = computers_data.iloc[::M, :]
average_price_systematic = systematic_sample['price'].mean()

# 4. Cluster Sampling (Assuming random clusters are chosen, and clusters are groups of premium/non-premium)
clusters = computers_data.groupby('premium')
cluster_sample = clusters.sample(n=sample_size // len(clusters), random_state=1)
average_price_cluster = cluster_sample['price'].mean()

# Print out the results
print(f"Average Price (Random Sampling): {average_price_random:.2f}")
print(f"Average Price (Stratified Sampling): {average_price_stratified:.2f}")
print(f"Average Price (Systematic Sampling): {average_price_systematic:.2f}")
print(f"Average Price (Cluster Sampling): {average_price_cluster:.2f}")
```

```python
#TASK 2.c:
# Group by 'premium' column and calculate the average price for each group
average_price_by_premium = computers_data.groupby('premium')['price'].mean()

# Output the results
print("Average Price of Premium and Non-Premium Computers in the Whole Data:")
print(average_price_by_premium)
# Set the sample size
sample_size = 100

# Simple random sampling without replacement
sampled_data = computers_data.sample(n=sample_size, random_state=1)

# Group by 'premium' and calculate the average price for each group in the sampled data
average_price_sampled_by_premium = sampled_data.groupby('premium')['price'].mean()

# Output the results
print("Average Price of Premium and Non-Premium Computers in the Sampled Data (Simple Random Sampling without Replacement):")
print(average_price_sampled_by_premium)
```

```python
#TASK 3.a:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming computers_data is a pandas DataFrame with a 'price' column

# Population (all computer prices in the dataset)
population_prices = computers_data['price'].dropna()

# Sample sizes to test
sample_sizes = [10, 50, 100, 500]

# Number of simulations (number of samples to take for each size)
num_simulations = 1000

# Function to simulate sampling and calculate sample means
def simulate_sample_means(sample_size, num_simulations):
    sample_means = []
    for _ in range(num_simulations):
        sample = population_prices.sample(n=sample_size, replace=False)
        sample_means.append(sample.mean())
    return sample_means

# Plotting the distributions of sample means for different sample sizes
plt.figure(figsize=(14, 10))

for i, sample_size in enumerate(sample_sizes, 1):
    sample_means = simulate_sample_means(sample_size, num_simulations)
    plt.subplot(2, 2, i)
    plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title(f"Sample Size = {sample_size}")
    plt.xlabel("Sample Mean Price")
    plt.ylabel("Frequency")
    plt.axvline(np.mean(population_prices), color='red', linestyle='dashed', linewidth=2, label='Population Mean')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
#TASK 3.c:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming computers_data is a pandas DataFrame with a 'price' column
np.random.seed(42)

# Let's assume a population of computer prices
population_prices = computers_data['price'].dropna()

# Step 1: Verify the population distribution (to check if the population is skewed or not)
plt.figure(figsize=(8, 6))
plt.hist(population_prices, bins=50, color='lightblue', edgecolor='black', alpha=0.7)
plt.title("Population Distribution of Computer Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# Step 2: Simulate the sampling distribution of the mean for different sample sizes
sample_sizes = [10, 30, 50, 100]
num_simulations = 1000

# Function to simulate sampling and calculate sample means
def simulate_sample_means(sample_size, num_simulations):
    sample_means = []
    for _ in range(num_simulations):
        sample = np.random.choice(population_prices, size=sample_size, replace=False)
        sample_means.append(sample.mean())
    return sample_means

# Step 3: Plot the sampling distribution of the sample mean for different sample sizes
plt.figure(figsize=(14, 10))

for i, sample_size in enumerate(sample_sizes, 1):
    sample_means = simulate_sample_means(sample_size, num_simulations)
    plt.subplot(2, 2, i)
    plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title(f"Sample Size = {sample_size}")
    plt.xlabel("Sample Mean Price")
    plt.ylabel("Frequency")
    plt.axvline(np.mean(population_prices), color='red', linestyle='dashed', linewidth=2, label='Population Mean')
    plt.legend()

plt.tight_layout()
plt.show()

# Step 4: Check mean and standard error of the sample means for different sample sizes
for sample_size in sample_sizes:
    sample_means = simulate_sample_means(sample_size, num_simulations)
    sample_mean = np.mean(sample_means)
    sample_std = np.std(sample_means)
    population_mean = np.mean(population_prices)
    population_std = np.std(population_prices)

    print(f"Sample Size = {sample_size}")
    print(f"Sample Mean = {sample_mean:.2f}, Population Mean = {population_mean:.2f}")
    print(f"Sample Std Dev = {sample_std:.2f}, Expected Standard Error = {population_std / np.sqrt(sample_size):.2f}\n")
```
