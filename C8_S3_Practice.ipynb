```python
#Task1:
import numpy as np
import scipy.stats as stats

# Given data
population_mean = 33.88  # average rent for designated industrial area in 2020
sample_mean = 32.18      # mean price per square foot from the sample
sample_size = 49         # number of samples
population_std_dev = 1.28  # standard deviation

# a) Null and alternative hypotheses
# H0: mue = 33.88 (mean rent has not changed)
# Ha: mue != 33.88 (mean rent has changed)

# Type of test: Two-tailed test

# b) Critical value for alpha = 0.05 (two-tailed)
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha / 2)
print("Critical value for alpha=0.05:", critical_value)

# c) Test statistic
# Calculate the standard error (SE)
standard_error = population_std_dev / np.sqrt(sample_size)

# Calculate the z-test statistic
z_test_statistic = (sample_mean - population_mean) / standard_error
print("Test Statistic (z):", z_test_statistic)

# d) Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_test_statistic)))  # two-tailed
print("P-value:", p_value)

# e) Draw conclusions
if abs(z_test_statistic) > critical_value:
    conclusion = "Reject the null hypothesis: There is a significant change in rental prices."
else:
    conclusion = "Fail to reject the null hypothesis: There is no significant change in rental prices."

print("Conclusion:", conclusion)

# Additionally, compare p-value with alpha
if p_value < alpha:
    conclusion_p_value = "Reject the null hypothesis: There is a significant change in rental prices."
else:
    conclusion_p_value = "Fail to reject the null hypothesis: There is no significant change in rental prices."

print("Conclusion based on p-value:", conclusion_p_value)
```

```python
#Task2:
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data
population_mean = 85           # Mean throwing distance before grip change
sample_mean = 87.5             # Mean throwing distance after grip change
sample_size = 32                # Number of throws
population_std_dev = 2.1        # Standard deviation

# a) Null and alternative hypotheses
# H0: mu <= 85 (mean distance has not increased)
# H1: mu > 85 (mean distance has increased)

# Type of test: One-tailed test (right-tailed)

# b) Critical value for alpha = 0.05 (one-tailed)
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha)
print("Critical value for alpha=0.05:", critical_value)

# c) Test statistic
# Calculate the standard error (SE)
standard_error = population_std_dev / np.sqrt(sample_size)

# Calculate the z-test statistic
z_test_statistic = (sample_mean - population_mean) / standard_error
print("Test Statistic (z):", z_test_statistic)

# d) Calculate the p-value
p_value = 1 - stats.norm.cdf(z_test_statistic)  # one-tailed
print("P-value:", p_value)

# e) Draw conclusions
if z_test_statistic > critical_value:
    conclusion = "Reject the null hypothesis: The grip change has significantly increased the throwing distance."
else:
    conclusion = "Fail to reject the null hypothesis: The grip change has not significantly increased the throwing distance."

print("Conclusion:", conclusion)

# Visualize the results
# Create a range of x values for the normal distribution
x = np.linspace(80, 95, 100)
y = stats.norm.pdf(x, population_mean, population_std_dev)

# Plot the normal distribution
plt.plot(x, y, label='Normal Distribution (mu=85, sigma=2.1)')
plt.fill_between(x, y, where=(x >= critical_value + population_mean), color='red', alpha=0.5, label='Critical Region')
plt.axvline(sample_mean, color='blue', linestyle='--', label='Sample Mean (87.5)')
plt.axvline(population_mean, color='green', linestyle='--', label='Population Mean (85)')
plt.title('Javelin Throw Distance Hypothesis Test')
plt.xlabel('Throw Distance (yards)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```
